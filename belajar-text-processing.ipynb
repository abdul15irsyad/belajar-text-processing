# Belajar Text Processing

```python
# import library
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
dataset = pd.read_csv('Data-Judul-Berita-Bulan-Januari-Tahun-2020.csv')
dataset.head()
```

```python
jumlah_kategori = dataset.groupby(['kategori']).size()
print(jumlah_kategori)
```

```python
stemmer = StemmerFactory().create_stemmer()
for index,data in dataset.iterrows():
    judul = data['judul_berita']
    normalize = judul.lower()
    normalize = re.sub("[^a-zA-Z ]","",normalize)
    normalize = stemmer.stem(normalize)
    dataset.loc[index,'normalize'] = normalize
dataset[['judul_berita','normalize']].head()
```

```python
list_word = set(stopwords.words('indonesian'))
removed = []
for index,data in dataset.iterrows():
    tokens = data['normalize'].split()
    filtered = [x for x in tokens if x not in list_word]
    removed = [x for x in tokens if x not in filtered]
    dataset.loc[index,'filtered'] = ' '.join(filtered)
    dataset.loc[index,'removed'] = ' '.join(removed)
dataset[['judul_berita','normalize','filtered','removed']].head()
```

```python
dataset.loc[:,['normalize','filtered','removed']].to_csv('result-text-processing.csv',index=False)
```
